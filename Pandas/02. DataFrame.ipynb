# 2. DataFrame
2차원데이터 (Series들의 모음)

## Data 준비
사전 (dict) 자료구조를 통해 생성

ex) 내 주위 친구들(나 포함) 8명에 대한 데이터 

```python
data = {
    '이름' : ['강병호', '김도윤', '김재열', '김형진', '이형준', '이병헌', '권동영', '최현우'],
    '학교' : ['포항고', '세명고', '제철공고', '포항고', '동지고', '제철고', '흥해공고', '세명고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
data
```

```python
# type(data)
data['이름'] # data['key']
```

```python
data['키']
```

## DataFrame 객체 생성

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df
```

## 데이터 접근

```python
df ['이름'] # df['가져오고자 하는 열 이름']
```

```python
df['키']
```

```python
df[['이름', '키']]
```

## DataFrame 객체 생성 (Index 지정)

```python
df = pd.DataFrame(data, index = ['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df
```

## DataFrame 객체 생성 (Column 지정)
data 중에서 원하는 column만 선택하거나 순서 변경 가능

```python
df = pd.DataFrame(data, columns = ['이름', '학교', '키'])
df
```

```python
df = pd.DataFrame(data, columns = ['이름', '키', '학교'])
df
```
