```python
import pandas as pd
data = {
    '이름' : ['강병호', '김도윤', '김재열', '김형진', '이형준', '이병헌', '권동영', '최현우'],
    '학교' : ['포항고', '세명고', '제철공고', '포항고', '동지고', '제철고', '흥해공고', '세명고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df = pd.DataFrame(data, index = ['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df
```

```python
df.index
```

## Index 이름 설정

```python
df.index.name = "지원번호"
df
```

## Index 초기화

```python
df.reset_index() # index column 으로 변화
```

```python
df.reset_index(drop = True) # 원래 쓰던 '지원번호' 인덱스 삭제
```

```python
df # 변화 x WHY?
# 위의 코드들은
# 실제 데이터에 반영 x 
```

```python
df.reset_index(drop = True, inplace = True) # 실제 데이터에 반영
```

```python
df
```

## Index 설정

```python
df.set_index('이름')
```

## Index 정렬
Index를 기준으로 오름차순, 내림차순 정렬

```python
df.sort_index() # 인덱스로 오름차순 정렬
```

```python
df.sort_index(ascending = False) # 인덱스로 내림차순 정렬
```
