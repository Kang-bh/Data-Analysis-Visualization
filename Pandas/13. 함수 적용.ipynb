## 데이터에 함수 적용 (apply)

```python
# 키 뒤에 cm 붙이는 함수
def add_cm(height):
    return str(height) + 'cm'
df['키'] = df['키'].apply(add_cm) # 키 데이터에 대해 add_cm 함수 호출한 결과 데이터 반영
```

```python
df
```

```python
def capitalize(lang):
    if pd.notnull(lang): # Nan이 아닌지
        return lang.capitalize() # string 내 제공, 첫 글자는 대문자 나머지는 소문자
    return lang
```

```python
df['SW특기'] = df['SW특기'].apply(capitalize)
```
