# 14. 그룹화
동일한 값을 가진 것들끼리 합쳐서 통ㅇ계 또는 평균 등의 값을 계산하기 위해 사용

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df.groupby('학교')
```

```python
df.groupby('학교').get_group('포항고')
```

```python
df.groupby('학교').get_group('세명고')
```

```python
df.groupby('학교').mean() # 계산가능한 데이터들의 평균값 계산
```

```python
df.groupby('학교').size() # 각 그룹의 크기 정보
```

```python
df.groupby('학교').size()['포항고'] # 학교로 그룹화한 뒤에 포항고에 해당하는 데이터의 수 확인
```

```python
df.groupby('학교')['키'].mean() # 학교로 그룹화한 뒤 키의 평균 데이터 확인
```

```python
df.groupby('학교')[['국어', '영어', '수학']].mean()
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2] # 학년 Column 추가
```

```python
df
```

```python
df.groupby(['학교', '학년']).mean() #학교별 학년별 평균데이터 알 수 있다.
```

```python
df.groupby('학년').mean()
```

```python
df.groupby('학년').mean().sort_values('키')
```

```python
df.groupby('학년').mean().sort_values('키', ascending = False)
```

```python
df.groupby(['학교', '학년']).sum() # 합 구하기
```

```python
df.groupby('학교')['SW특기'].count() # 학교로 그룹화한 후 각 학교 별 SW특기 개수 구하기
```

```python
df.groupby('학교')[['이름', 'SW특기']].count()
```

```python
school = df.groupby('학교')
school['학년'].value_counts() # 학교로 그룹화 한 뒤에 학년별로 학생 수를 가져옴
```

```python
school['학년'].value_counts().loc['포항고'] # 포항고에서만의 그룹화한 후 그 학년별의 학생 수 가져옴
```

```python
school['학년'].value_counts(normalize = True).loc['포항고'] # 학생들의 수 데이터를 비율로 가져온다.
```
