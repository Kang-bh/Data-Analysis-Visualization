# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인지 Boolean 형식으로 표시
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt] # filter 를 역으로 적용
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185, '수학'] # 키가 185 이상인 학생들의 수학 데이터
```

```python
df.loc[df['키'] >= 185, ['이름', '수학', '과학']] # 키가 185이상인 학생들의 이름, 수학, 과학 데이터를 가져온다
```

## 다양한 조건

### & 그리고

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '포항고')] # 키가 185이상인 포항고 학생 데이터 가져옴
```

### | 또는

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)] # 키가 170보다 작거나 200보다 큰 학생들의 데이터
```

### str 함수

```python
filt = df['이름'].str.startswith('김') # '김씨 성을 가진 사람'
df[filt]
```

```python
filt = df['이름'].str.contains('병') # '병' 이 들어가는 사람 뽑음
df[filt]
```

```python
df[~filt] # '병'이 들어가는 사람 제외
```

```python
langs = ["Python", "Java"]
filt = df["SW특기"].isin(langs) # SW특기가 Python OR JAVA
df[filt]
```

```python
langs = ["python", "java"]
filt = df["SW특기"].str.lower().isin(langs)
df[filt]
```

```python
filt = df["SW특기"].str.contains('Java')
df(filt)
```

```python
df['SW특기'].str.contains('Java', na = True) # Nan data TRUE로
```

```python
df['SW특기'].str.contains('Java', na = False) # Nan data False로
```

```python
filt = df["SW특기"].str.contains('Java', na = False)
df[filt]
```
