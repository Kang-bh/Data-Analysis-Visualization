## 데이터 채우기 fillna

```python
df.fillna('') # Nan 데이터를 빈 칸으로 채움
```

```python
df.fillna('없음') # Nan 데이터를 없음으로 채움
```

```python
import numpy as np
df['학교'] = np.nan # 학교 데이터 전체를 Nan으로 채움
df
```

```python
df.fillna('모름')
```

```python
df
```

```python
df.fillna('모름', inplace = True)
df
```

## 데이터 제외하기 dropna

```python
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df.dropna(inplace = True) # Nan 을 포함하는 데이터 삭제
```

```python
df
```

```python
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

- axis : index or columns
- how: any or all

```python
df.dropna(axis = 'index', how = 'any') # NaN 데이터가 하나라도 있는 row 삭제
```

```python
df.dropna(axis = 'columns') # Nan이 하나라도 있는 columns 삭제
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axios = 'columns', how = 'all') # 데이터 전체가 NaN인 Columns 만 삭제
```
