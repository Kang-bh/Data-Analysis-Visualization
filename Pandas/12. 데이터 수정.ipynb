# 12. 데이터  수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'제철고':'포항제철고', '동지고':'동지여고'})
```

```python
df['학교'].replace({'제철고':'포항제철고'}, inplace = True)
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교데이터 + 등학교
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['수학'] + df['영어'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail' # 결과 Column 추가하고 전체 데이터 Fail로 초기화
df
```

```python
df.loc[df['총합'] >= 400, '결과'] = 'Pass' # 총합이 400보다 큰 데이터의 결과 Pass로 업데이트
```

```python
df
```

## Column 삭제

```python
df.drop(columns = ['총합'])  # '총합' Coulmn 삭제
```

```python
df.drop(columns=['국어', '영어', '수학'])
```

```python
## inplace 해야 실제 데이터에 반영
```
