```python
labels = ['강병호', '이병헌', '김세환'] # 이름
values = [177, 180, 183 ] # 키

plt.barh(labels, values) # 오른쪽
plt.xlim(175, 185)
plt.show()
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch("/") # ////
bar[1].set_hatch("x") # xxxx
bar[2].set_hatch("..") # ....
```

참고자료 : https://matplotlib.org/devdocs/gallery/shapes_and_collections/hatch_style_reference.html

```python
bar = plt.bar(labels, values)
plt.ylim(175, 185)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height()+ 0.5, values[idx], ha = 'center', color = 'blue')
```
