# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시 마이너스 글자 깨짐 현상 방지
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'C++', 'Javascript', 'Java', 'C#', 'ETC']

plt.pie(values, labels = labels, autopct = '%.1f')
plt.show()
```

```python
values = [1, 1, 1, 1, 1, 1]
labels = ['Python', 'C++', 'Javascript', 'Java', 'C#', 'ETC']

plt.pie(values, labels = labels, autopct = '%.1f', startangle= 90, counterclock = False) # percentage 값으로 변환, 시계방향
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'C++', 'Javascript', 'Java', 'C#', 'ETC']
# explode = [0.2, 0.1, 0, 0, 0, 0]
explode = [0.05] * 6

plt.pie(values, labels = labels, explode = explode, autopct = '%.1f')
plt.show()
```

```python
plt.pie(values, labels = labels, explode = explode)
plt.legend()
plt.show()

# 범례가 가리는 문제
```

```python
plt.pie(values, labels = labels, explode = explode)
plt.title("언어별 선호도")
plt.legend(loc = (1.2, 0.3), title = "언어") # x = 1.2 y = 0.3 만큼 띄우기, 제목: 언어별 선호도
plt.show()
```
